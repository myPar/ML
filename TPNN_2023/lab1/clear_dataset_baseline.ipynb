```python
import pandas as pd

input_file_name = "../datasets/ID_data_mass_18122012.xlsx"
output_file_name = "../datasets/ID_data_mass_18122012_cleared.xlsx"
```

### Check dataframe

```python
df = pd.read_excel(input_file_name, sheet_name='VU')
header = df.columns.to_numpy()
print(header)
df
```

### Setting up correct header

```python
### setting name to two dropped columns
dropped_columns = ['номер', 'дата']
df.iloc[0,0] = dropped_columns[0]
df.iloc[0,1] = dropped_columns[1]

### setting header as 1'st row
df.columns = df.iloc[0]
df
```

```python
### drop redundant rows
df.drop(0, inplace=True)
df.drop(1, inplace=True)
df
```

```python
# update index
df.reset_index(drop=True)
df
```

### Replace '-' on Nan

```python
def replace_skip(string: str):
    if string == '-':
        return float('nan')
    else:
        return string

df = df.applymap(replace_skip)
df
```

### Concatenate two 'КГФ' attributes

```python
### rename one 'КГФ' column
attribute_count = len(df.columns)
df.columns.values[attribute_count-2] = 'КГФ_'
df
```

```python
### divide 'КГФ_' on 1000 to be the same dimention as 'КГФ'
df['КГФ_'] = df['КГФ_'].apply(lambda x: x / 1000)
df
```

```python
### removes column 'КГФ_' summarize it with 'КГФ' previously
def sum_columns(dataframe, src_col_name, dst_col_name):
    dataframe[dst_col_name] = dataframe[src_col_name].fillna(0) + \
                              dataframe[dst_col_name].fillna(0)
    ### fill nan back (there were no zeroes before)
    dataframe[dst_col_name] = dataframe[dst_col_name].apply(lambda x: float('nan') if x == 0 else x)
    dataframe.drop(columns=[src_col_name], axis=1, inplace=True)

sum_columns(df, 'КГФ_', 'КГФ')
df
```

### Check type of attributes

```python
df_types = df.dtypes
df_types
```

```python
dict_df = df.to_dict(orient='list')
attributes = dict_df.keys()

print(dict_df.keys())
print("attribute count=" + str(len(attributes)))
print(dict_df['Tлин'])
```

```python
object_attributes = df.select_dtypes(include='object').columns.to_list()
object_attributes
```

```python
### check object attribute values
print(dict_df[object_attributes[1]])
```

```python
def try_cast_float(item: str):
    try:
        return float(item)
    except ValueError:
        return float('nan')

### replace string items on nan in object attributes
for o_attribute in object_attributes:
    for i in range(len(dict_df[o_attribute])):
        item = dict_df[o_attribute][i]

        if isinstance(item, str):
            dict_df[o_attribute][i] = try_cast_float(item)

print(dict_df[object_attributes[1]])
```

### Save cleared dataframe

```python
cleared_df = pd.DataFrame.from_dict(dict_df)
cleared_df
```
