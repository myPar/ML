### Plot functions

```python
def plot_loss(loss_list):
  # plot loss graphic
  x = np.arange(1, len(loss_list) + 1)
  y = np.array(loss_list)

  assert len(x) == len(y)

  fig = px.line(x=x, y=y, labels={'x':'epoch', 'y':'loss'})
  fig.show()

def plot_approximation(df, attribute_name, predicted_data, mode: str):
    assert attribute_name in df.columns
    assert df.shape[0] == predicted_data.shape[0]

    comparison_graphic = go.Figure()

    x = np.arange(1, predicted_data.shape[0] + 1)
    y_actual = df[attribute_name].to_numpy()
    y_predicted = predicted_data

    comparison_graphic.add_trace(go.Scatter(x=x, y=y_actual, mode=mode, name=attribute_name + ' actual graphic'))
    comparison_graphic.add_trace(go.Scatter(x=x, y=y_predicted, mode=mode, name=attribute_name + ' predicted'))

    comparison_graphic.update_layout(
    height= 400,
    width = 1200,
    title_text=attribute_name
    )
    return comparison_graphic
```

### Help reused functions

```python
def get_predicted_data(input_data, net: Net):
    predicted_data = []

    for item in input_data:
        predicted_item = net.calc_output(item)
        predicted_data.append(predicted_item[0])

    predicted_data = np.array(predicted_data)

    return predicted_data
```

### Prepare train data

```python
### get formatted train data
output_attributes = ['КГФ']
input_attributes = np.setdiff1d(np.array(kgf_selected_attributes), output_attributes)
print(input_attributes)
print(output_attributes)
```

```python
df_cleared[input_attributes]
```

```python
df_cleared[output_attributes]
```

```python
input_train_data = df_cleared[input_attributes].to_numpy()
output_train_data = df_cleared[output_attributes].to_numpy()
print(input_train_data.shape)
print(output_train_data.shape)
```

### Build and train net

```python
net = Net()
net.add_layer(Dense(90, sigmoid))
net.add_layer(Dense(1, sigmoid))
net.init_net(init_strategy=default_init, input_shape=(4,))
net.print_net_config(level=ConfigLevel.MEDIUM)
```

```python
net.train(input_train_data=input_train_data, target_train_data=output_train_data, optimizer=Adam(), learning_rate=0.0001, epoch_count=400, loss=mse)
```

### Plot approximation and loss

```python
plot_loss(net.loss_list)
```

```python
predicted_data = get_predicted_data(input_train_data, net)

plot_approximation(df_cleared, 'КГФ', predicted_data, mode='lines+markers')
```
