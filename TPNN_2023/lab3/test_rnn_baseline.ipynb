```python
from TPNN_2023.lab2.Perceptron import ConfigLevel
from TPNN_2023.lab2.metrics import mse
from TPNN_2023.lab2.tools import Adam
from TPNN_2023.lab3.rnn import RNNnet, RNNlayer, LOSS_TYPE
from TPNN_2023.lab2.activation_functions import *
import pandas as pd
import numpy as np
import plotly.express as px
import plotly.figure_factory as ff
import plotly.graph_objects as go
```

```python
data = np.random.normal(size=(1000, 4))
data
```

```python
df = pd.DataFrame(data=data, columns=['a', 'b', 'c', 'd'])
df
```

```python
input_train_data = df[['a', 'b', 'c']].to_numpy()
output_train_data = df[['d']].to_numpy()

timestamps = 20
batch_count = data.shape[0] // timestamps

input_train_data = input_train_data.reshape((batch_count, timestamps, 3, 1))
output_train_data = output_train_data.reshape(batch_count, timestamps, 1, 1)
input_train_data
```

```python
rnn_net = RNNnet()
rnn_net.add_layer(RNNlayer(timestamps=timestamps, neuron_count=15, output_vector_shape=(1, 1), activation=th, input_vector_shape=(3, 1)))
```

```python
rnn_net.calc_output(input_train_data[0])
```

```python
rnn_net.set_loss(LOSS_TYPE.MAE)
rnn_net.train(input_train_data=input_train_data, target_train_data=output_train_data, optimizer=Adam(), epoch_count=100, lr=0.00000001)
```

```python
predictions = rnn_net.predict(input_train_data)
predictions.shape
```

```python
predictions = predictions.reshape((batch_count, timestamps, 1))
predictions = predictions.flatten()
predictions
```

```python
def plot_approximation(predicted_data, dst_data, attribute_name, mode: str):
    assert predicted_data.shape == dst_data.shape

    comparison_graphic = go.Figure()

    x = np.arange(1, predicted_data.shape[0] + 1)
    y_actual = dst_data
    y_predicted = predicted_data

    comparison_graphic.add_trace(go.Scatter(x=x, y=y_actual, mode=mode, name=attribute_name + ' actual graphic'))
    comparison_graphic.add_trace(go.Scatter(x=x, y=y_predicted, mode=mode, name=attribute_name + ' predicted'))

    comparison_graphic.update_layout(
    height= 1000,
    width = 4000,
    title_text=attribute_name
    )
    return comparison_graphic
```

```python
base_dir = '../plots/lab3/'
comparison_graphic = plot_approximation(predictions, output_train_data.flatten(), 'd','markers')
comparison_graphic.write_image(base_dir + "rnn_test.pdf")
```
