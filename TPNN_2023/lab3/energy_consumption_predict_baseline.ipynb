```python
from TPNN_2023.lab2.Perceptron import ConfigLevel
from TPNN_2023.lab2.metrics import mse
from TPNN_2023.lab2.tools import Adam
from TPNN_2023.lab3.rnn import RNNnet, RNNlayer
from TPNN_2023.lab2.activation_functions import *
```

```python
rnn_net = RNNnet()
rnn_net.add_layer(RNNlayer(timestamps=10, neuron_count=10, output_vector_shape=(8, 1), activation=th, input_vector_shape=(12,1)))
rnn_net.add_layer(RNNlayer(timestamps=10, neuron_count=14, output_vector_shape=(10, 1), activation=th, input_vector_shape=(8,1)))
```

```python
rnn_net.print_net_config(ConfigLevel.HIGH)
```

```python
input_tensor = np.ones((10, 12, 1))

rnn_net.calc_output(input_tensor)
```

```python
input_train_data = np.ones((1, 10, 12, 1)) * 2
target_train_data = np.ones((1, 10, 10, 1))
rnn_net.train(input_train_data=input_train_data, target_train_data=target_train_data, optimizer=Adam(), loss=mse, epoch_count=2, lr=1)
```
