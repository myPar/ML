```python
import numpy as np
import pandas as pd
input_file_name = "../data_sets/lab1.xlsx"
output_file_name = "../data_sets/lab1_prepared.xlsx"
```

```python
drop_columns = ["номер", "дата"]
df = pd.read_excel(input_file_name, sheet_name='VU')
header = df.columns.to_numpy()
df
```

```python
# set new header (line 2)
df.iloc[0,0] = drop_columns[0]
df.iloc[0,1] = drop_columns[1]
df.columns = df.iloc[0]
df
```

```python
# drop redurant rows
df.drop(0, inplace=True)
dim = df.drop(1, inplace=True)
df
```

```python
df = df.reset_index(drop=True)
df
```

```python
# replace all '-' on Nan
def replace_skip(string: str):
    if string == '-':
        return np.nan
    else:
        return string

filled_df = df.applymap(replace_skip)
filled_df
```

**remove first 'КГФ' column**

```python
# remove it with boolean mask
df = filled_df
mask = np.ones(df.shape[1], dtype=bool)
mask[len(mask)-2] = False

df = df.loc[:, mask]
df
```

**remove rows where G_total and КГФ are nan**

```python
classes = ['G_total', 'КГФ']
df = df[(df[classes[0]].notna()) | (df[classes[1]].notna())]
df = df.reset_index(drop=True)
df
```

**removing 'str' items**

```python
#convert df to dict view
dict_df = df.to_dict(orient='list')
print(dict_df.keys())
print(dict_df["номер"])
```

```python
# remove 'str' items from dataframe dict
def fill_str(data):
    for i in range(len(data)):
        item = data[i]

        if type(item) == str:
            data[i] = np.nan


for key in dict_df:
    if key != drop_columns[0] and key != drop_columns[1]:
        fill_str(data=dict_df[key])

#save dict as dataframe back
df = pd.DataFrame.from_dict(dict_df)
df
```

**save prepared dataset**

```python
df.to_excel(output_file_name, index=False)
```
