Missing data

```python
import numpy as np
import pandas as pd
from numpy.random import randn
```

```python
# drop nan values
df = pd.DataFrame(randn(5,4), index=[1,2,3,4,5], columns=['A', 'B', 'C', 'D'])
df['A'][3] = np.nan
df['D'][4] = np.nan
df
```

```python
# drop all rows with nan
df.dropna()
```

```python
# drop all columns with nan
df.dropna(axis=1)
```

```python
# threshold using
df['D'][1] = np.nan
df
```

```python
df.dropna(axis=1, thresh=4)
```

```python
# fill nan values
df.fillna(value=1)
```

```python
# fill nan by function of items
df['D'].fillna(value=df['D'].mean(), inplace=True)
df
```
