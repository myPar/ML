Matplotlib (axes)

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0, 5, 11)
x
```

```python
y = x ** 2
y
```

FUNCTIONAL way

```python
# plot graphic
plt.plot(x, y)
plt.xlabel('X label')
plt.ylabel('Y label')
plt.title('My Graphic')
plt.show()
```

```python
# multiple plots
plt.subplot(1, 2, 1)
plt.plot(x, y, 'red')

plt.subplot(1, 2, 2)
plt.plot(y, x, 'blue')
plt.show
```

OO method

```python
fig = plt.figure()  # making a 'figure' objects
axes = fig.add_axes([0.1,0.1, 0.8, 0.8])
axes.plot(x,y)

axes.set_xlabel('X label')
axes.set_ylabel('Y label')

fig.show()
```

```python
# several axes on one graphic
fig = plt.figure()
axes1 = fig.add_axes([0.1, 0.1, 1, 1])
axes2 = fig.add_axes([0.2, 0.52, 0.5, 0.5])

axes1.plot(x, y, 'red')
axes2.plot(y, x, 'green')

axes1.set_title('LARGE plot')
axes2.set_title('SMALL plot')

fig.show()
```

OO subplots

```python
fix, axes = plt.subplots(nrows=2, ncols=2)  # returns one figure and axes array
axes = np.squeeze(axes.reshape((1, 4)), axis=0)

colours = ['red', 'green', 'blue', 'black'] # also available RGB hex-code

for ax, colour in zip(axes, colours):
  ax.set_xlabel('X_label')
  ax.set_ylabel('Y_label')
  ax.plot(x, y, colour)
plt.tight_layout()
fig.show()  
```

DPI and figure size

```python
# set figure size
fix, axes = plt.subplots(nrows=2, ncols=1, figsize=(8,6))
axes[0].plot(x, y, 'r', label='x squared')
axes[0].legend(loc=0)

axes[1].plot(y, x,'b', label='x square root')
axes[1].legend(loc=0)
plt.tight_layout()
fig.show()
```

```python
# save figure as picture with set dpi
fig.savefig("image.png", dpi=200)
```

Customisation

```python
fix, axes = plt.subplots(nrows=1, ncols=2, figsize=(10,5))
ax1 = axes[0]
ax2 = axes[1]

x = np.linspace(0, 10, 30)
y = x**(1/2)

# custom colour and thickness
ax1.plot(x, y, color='#FA72F6', linewidth=6)

# custom transparency
ax2.plot(x, y, color='#FA72F6', linewidth=6, alpha=0.5) #'linewidth'='lw'
fig.show()
```

Line styles

```python
# https://matplotlib.org/stable/gallery/lines_bars_and_markers/linestyles.html
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(10,5))

ax1 = axes[0]
ax2 = axes[1]

ax1.plot(x, y, linestyle=(0, (5, 10, 2, 7)))  #'linestyle'='ls'
ax2.plot(x, y, linestyle='-.')
fig.show()
```

Mark points

```python
fig = plt.figure()
ax = fig.add_axes([0, 0, 1, 1])
ax.plot(x, y, marker='o', markersize='10', markerfacecolor='yellow', markeredgecolor='black', markeredgewidth=3)
fig.show()
```

Bound graphics

```python
ax.set_xlim([2, 4])
ax.set_ylim([1, 2])
```
