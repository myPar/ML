# Plots

destribution plots

```python
import seaborn as sb
%matplotlib inline
```

```python
tips = sb.load_dataset('tips')
tips.head()
```

```python
# distribution plot of total_bill attribute:
sb.displot(tips['total_bill'], kde=False, bins=30)
```

```python
# plot of two variables
sb.jointplot(x='total_bill', y='tip', data=tips)
```

```python
sb.jointplot(x='total_bill', y='tip', data=tips, kind='reg')
```

```python
sb.jointplot(x='total_bill', y='tip', data=tips, kind='kde')
```

```python
# pair plots
sb.pairplot(tips, hue='sex')
```

categorical plots

```python
import numpy as np
```

```python
sb.barplot(x='sex', y='total_bill', data=tips)
```

```python
sb.countplot(x='sex', data=tips)
```

```python
sb.boxplot(x='day', y='total_bill', data=tips, hue='smoker')
```

```python
sb.violinplot(x='day', y='total_bill', hue='sex', data=tips, split=True)
```

```python
sb.stripplot(y='day', x='total_bill', data=tips)
```

```python
sb.factorplot(x='day', y='total_bill', data=tips, kind='violin')
```

# Matrices

```python
import seaborn as sb
%matplotlib inline
tips = sb.load_dataset('tips')
flights = sb.load_dataset('flights')
tips.head()
```

```python
flights.head()
```

```python
tips_corr = tips.corr()
tips_corr
```

```python
sb.heatmap(tips_corr, annot=True, cmap='coolwarm')
```

```python
ft = flights.pivot_table(index='month', columns='year', values='passengers')
ft
```

```python
sb.heatmap(ft, cmap='magma', linecolor='white', linewidth=1)
```

```python
sb.clustermap(ft, cmap='magma', standard_scale=0)
```

Grids

```python
iris = sb.load_dataset('iris')
iris.head()
```

```python
sb.pairplot(iris)
```

```python
pg = sb.PairGrid(iris)
pg.map_diag(sb.histplot)
pg.map_upper(sb.scatterplot)
pg.map_lower(sb.kdeplot)
```

```python
pg = sb.FacetGrid(data=tips, col='time', row='smoker')
pg.map(sb.histplot, 'total_bill')
```
