```python
import numpy as np
import pandas as pd
from numpy.random import randn
```

```python
df = pd.DataFrame(randn(5,4), index=[1,2,3,4,5], columns=["A", 'B', 'C', 'D'])
df
```

```python
df["A"]
```

```python
df[['A', 'C']]
```

```python
# reset the existing column or create a new one
print(df)
df['A'] = df['A'] + df['B']
print(df)
```

```python
# remove column from existing df (inplace=True)
df.drop('A',axis=1, inplace=True)
df
```

```python
df.drop(5, inplace=True)
df
```

```python
# select rows
df.loc[4]
```

```python
df.iloc[0]
```

```python
# select set of rows and columns
df.loc[1, 'B']
```

```python
df.loc[[1, 2], ['B', 'C']]
```

Conditional selection

```python
df > 0
```

```python
bool_df = df > 0
# mask on original df
df[bool_df]
```

```python
# not satisfying the condition rows are removed
df[df['B'] > 0]
```

```python
# new data frame
df = pd.DataFrame(randn(5,6), index=[1,2,3,4,5], columns=["A", 'B', 'C', 'D', 'E', 'F'])
df
```

```python
# multiple conditions
df[(df['A'] > 0) & (df['D'] > 0)]
```

```python
df[(df['A'] > 0) | (df['D'] > 0)]
```

indexies

```python
# adding default index and set old as a column
df.reset_index()
```

```python
# avoid adding old indexies as a column
df.reset_index(drop=True)
```

```python
df["keys"] = ['a', 'b', 'c', 'd', 'e']
df
```

```python
# set indexies using existing column
df.set_index("keys")
```

index levels

```python
outer_index = ['A', 'A', 'A', 'B', 'B', 'B']
inner_index = [1,2,3,1,2,3]
hier_index = list(zip(outer_index, inner_index))
hier_index
```

```python
# create multi-index
hier_index = pd.MultiIndex.from_tuples(hier_index)
hier_index
```

```python
df = pd.DataFrame(randn(6, 3), hier_index, ['A', "B", "C"])
df
```

```python
# access throw multiple index
df.loc['A'].loc[[1, 2]]
```

```python
# set index names
df.index.names = ['Group', 'Number']
df
```

```python
# get value
df.loc['B'].loc[2]['C']
```

```python
# cross-section
df.xs('A')
```

```python
# the same with 'loc'
df.loc['A']
```

```python
# get all '1' from all outer indexies
df.xs(1, level='Number')
```
